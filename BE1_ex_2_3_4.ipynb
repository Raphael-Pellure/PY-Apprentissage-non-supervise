```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

```

X est la matrice des donn´ees (de taille n ˆ p), k le nombre de clusters souhait´e, M une matrice contenant les centro¨ıdes initiaux (de taille k ˆp), et maxit d´efinit le nombre maximal d’it´erations accept´e. Lorsque l’utilisateur ne fournit pas la matrice M des centro¨ıdes initiaux (donc M is None est vraie), l’algorithme devra la construire en choisissant al´eatoirement (avec probabilit´e uniforme) k donn´ees parmi les n fournies.

```python
def kmeans(X, k, M=None, maxit=100):
    n, p = X.shape
    if M is None:
        random_indices = np.random.choice(n, size = k, replace=False)
        M = X[random_indices]

    for it in range(maxit):

        # --- MAJ DE Z --- #
        D = np.linalg.norm(X[:, np.newaxis, :] - M[np.newaxis, :, :], axis=2)

        # assigne les clusters  
        C = np.argmin(D, axis=1)

        Z_new = np.zeros((n, k), dtype=int)

        Z_new[np.arange(n), C] = 1
        
        # --- MAJ DE M --- #
        
        M_new = np.zeros_like(M)

        M_new = Z_new.T @ X
        counts = Z_new.sum(axis=0)[:, np.newaxis]
        M_new = M_new / counts

        # for j in range(k):
        #     count = Z_new[:, j].sum()
        #     if count > 0:
        #         M_new[j] = (Z_new[:, j][:, None] * X).sum(axis=0) / count
        #     else:
        #         M_new[j] = M[j]

        # convergence check
        if np.allclose(M, M_new):
            break
        M = M_new

    return M, C
```

```python
n1 = 50
m_p = [1, 1]

n2 = 80
m_m = [-1,-1]

sigma_2 = 5
X1 = norm.rvs(loc=m_p, scale=np.sqrt(sigma_2), size=(n1, 2))
X2 = norm.rvs(loc=m_m, scale=np.sqrt(sigma_2), size=(n2, 2))

X = np.vstack((X1, X2))
n, p = X.shape

# question 2 c)
# A1 et A2 sont des matrices tirées iid gaussiennes
A1 = norm.rvs(size=(p, p)) 
A2 = norm.rvs(size=(p, p)) 

# normal moyenne m_p et covariance A1*A1^T
X1_c = norm.rvs(loc=m_p, scale=1, size=(n1, p)) @ A1
# normal moyenne m_m et covariance A2*A2^T
X2_c = norm.rvs(loc=m_m, scale=1, size=(n2, p)) @ A2

X = np.vstack((X1_c, X2_c))
n, p = X.shape

k = 2

M, C = kmeans(X, k)

y_true = np.concatenate((np.ones(n1), np.zeros(n2)))

# Nombre de bonnes classifications
nb1 = np.sum(C == y_true)
nb2 = np.sum((1 - C) == y_true)  # cas où les labels sont inversés
nb_bonnes = max(nb1, nb2)
print(f"Nombre de bonnes classifications: {nb_bonnes} / {n}")


# taux d'erreur de classification
taux_erreur = (n - nb_bonnes) / n 
print(f"Taux d'erreur de classification: {taux_erreur*100:.2f}%")

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=C)
plt.scatter(M[:, 0], M[:, 1], c='red', marker='X', label='Centroids')
plt.title('K-means')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()

```

```python
N = 500
# 0.3 0.4 ... 5
sigma_2_values = np.arange(3, 51)/10

avg_taux_erreur_liste = []
ecart_type_liste = []

n1 = 50
m_p = [1, 1]

n2 = 80
m_m = [-1,-1]

for sigma_2 in sigma_2_values:
    taux_erreur = []
    ecart_type = []

    for i in range(N):

        X1 = norm.rvs(loc=m_p, scale=np.sqrt(sigma_2), size=(n1, 2))
        X2 = norm.rvs(loc=m_m, scale=np.sqrt(sigma_2), size=(n2, 2))

        X = np.vstack((X1, X2))

        n, p = X.shape
        k = 2

        M, C = kmeans(X, k)

        y_true = np.concatenate((np.ones(n1), np.zeros(n2)))

        # Nombre de bonnes classifications
        nb1 = np.sum(C == y_true)
        nb2 = np.sum((1 - C) == y_true)  # cas où les labels sont inversés
        nb_bonnes = max(nb1, nb2)

        # taux d'erreur de classification
        taux_erreur.append((n - nb_bonnes) / n)

    # moyenne taux d'erreur
    avg_taux_erreur = np.mean(taux_erreur)
    avg_taux_erreur_liste.append(avg_taux_erreur)

    # ecart type
    ecart_type = np.std(taux_erreur)
    ecart_type_liste.append(ecart_type)
    #print(f"Sigma_2: {sigma_2}, Moyenne taux d'erreur: {avg_taux_erreur:.2f}, Ecart type: {ecart_type:.2f}")


# plot avg taux d'erreur vs sigma_2 and ecart type as error bars
plt.figure(figsize=(10,6))
plt.errorbar(sigma_2_values, avg_taux_erreur_liste, yerr=ecart_type_liste, fmt='-o', ecolor='red', capsize=5)
plt.title('Taux d\'erreur moyen en fct de Sigma_2')
plt.xlabel('Sigma_2')
plt.ylabel('Taux d\'erreur moyen')
plt.grid(True)
plt.show()
```

```python
#knn++
def kmeansppinit(X, k, maxit=100):
    n, p = X.shape

    # Initialisation des centres avec k-means++
    M = np.zeros((k, p))
    M[0] = X[np.random.choice(n)]

    for j in range(1, k):
        D_i = np.min(np.linalg.norm(X[:, np.newaxis, :] - M[np.newaxis, :j, :], axis=2), axis=1)
        probs = D_i**2 / np.sum(D_i**2)
        M[j] = X[np.random.choice(n, p=probs)]

    return M
```

```python
# kmedoids
def kmedoids(X, k, dissim, m=None, maxit=100):
    n, p = X.shape
    if m is None:
        random_indices = np.random.choice(n, size = k, replace=False)
        m = X[random_indices]

    for it in range(maxit):

        # --- MAJ DE Z --- #
        D = dissim(X, m)

        # assigne les clusters  
        C = np.argmin(D, axis=1)

        Z_new = np.zeros((n, k), dtype=int)

        Z_new[np.arange(n), C] = 1
        
        # --- MAJ DE m --- #
        
        m_new = np.zeros_like(m)

        for j in range(k):
            indices_cluster = np.where(C == j)[0]
            if len(indices_cluster) > 0:
                cluster_points = X[indices_cluster]
                distances = dissim(cluster_points, cluster_points)
                total_distances = distances.sum(axis=1)
                medoid_index = np.argmin(total_distances)
                m_new[j] = cluster_points[medoid_index]
            else:
                m_new[j] = m[j]

        # convergence check
        if np.allclose(m, m_new):
            break
        m = m_new

    return m, C

```

```python
n1 = 13
m_p = [1/2, 1/2]

n2 = 20
m_m = [-1/2,-1/2]

sigma_2 = 0.3
X1 = norm.rvs(loc=m_p, scale=np.sqrt(sigma_2), size=(n1, 2))
X2 = norm.rvs(loc=m_m, scale=np.sqrt(sigma_2), size=(n2, 2))

X = np.vstack((X1, X2))
n, p = X.shape

# question 2 c)
# A1 et A2 sont des matrices tirées iid gaussiennes
A1 = norm.rvs(size=(p, p)) 
A2 = norm.rvs(size=(p, p)) 

# normal moyenne m_p et covariance A1*A1^T
X1_c = norm.rvs(loc=m_p, scale=1, size=(n1, p)) @ A1
# normal moyenne m_m et covariance A2*A2^T
X2_c = norm.rvs(loc=m_m, scale=1, size=(n2, p)) @ A2

X = np.vstack((X1_c, X2_c))

# rajout des outliers a la classe m_p
x_34 = [5, 5]
x_35 = [6, 6]
X = np.vstack((X, x_34, x_35))
n, p = X.shape
```

```python
# kmeans
k = 2
M = kmeansppinit(X, k)
M, C = kmeans(X, k, M)

# kmedoids
def dissim_euclidienne(X, Y):
    return np.linalg.norm(X[:, np.newaxis, :] - Y[np.newaxis, :, :], axis=2)

M_medoid, C_medoid = kmedoids(X, k, dissim_euclidienne)

# true labels et outliers
y_true = np.concatenate((np.ones(n1), np.zeros(n2)))
y_true = np.append(y_true, [1, 1])  # outliers appartiennent a la classe m_p

# Nombre de bonnes classifications kmeans
nb1 = np.sum(C == y_true)
nb2 = np.sum((1 - C) == y_true) 
nb_bonnes = max(nb1, nb2)
print(f"Nombre de bonnes classifications kmeans: {nb_bonnes} / {n}")

#Nombre de bonnes classifications kmedoids
nb1_medoid = np.sum(C_medoid == y_true)
nb2_medoid = np.sum((1 - C_medoid) == y_true)
nb_bonnes_medoid = max(nb1_medoid, nb2_medoid)
print(f"Nombre de bonnes classifications kmedoids: {nb_bonnes_medoid} / {n}")


# taux d'erreur de classification
taux_erreur = (n - nb_bonnes) / n 
print(f"Taux d'erreur de classification: {taux_erreur*100:.2f}%")

taux_erreur_medoid = (n - nb_bonnes_medoid) / n
print(f"Taux d'erreur de classification kmedoids: {taux_erreur_medoid*100:.2f}%")

plt.figure(figsize=(6,6))
plt.scatter(X[:, 0], X[:, 1], c=C, marker='o')
plt.scatter(X[:, 0], X[:, 1], c=C_medoid, marker='o')

plt.scatter(M[:, 0], M[:, 1], c='red', marker='X', label='Centroids')
plt.scatter(M_medoid[:, 0], M_medoid[:, 1], c='blue', marker='X', label='Medoids')

plt.title('K-means and K-medoids')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()

```
